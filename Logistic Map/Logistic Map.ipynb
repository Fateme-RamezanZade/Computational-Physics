1. Logistic Map Bifurcation Diagram

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi']=150
plt.rcParams['savefig.dpi']=100

def f(r,n,x_0):
    x=x_0
    y=np.zeros(100)
    for i in range(n):
        x=4*r*(x*(1-x))
        if i>=n-100:
          y[n-i-1]=x
    return y

r_values=np.arange(0,1,0.001)
N=len(r_values)
f_values=np.zeros((N,100))
x_0 = 0.2
for i in range(N):
    r=r_values[i]
    f_values[i]=f(r,1000,x_0)
plt.plot(r_values,f_values,'bo',markersize=0.3)
plt.title("Logistic map bifurcation diagram")
plt.xlabel("r")
plt.ylabel("x*")
plt.show()

r_values=np.arange(0.6,1,0.001)
N=len(r_values)
f_values=np.zeros((N,100))
x_0 = 0.2
for i in range(N):
    r=r_values[i]
    f_values[i]=f(r,1000,x_0)
plt.plot(r_values,f_values,'bo',markersize=0.3)
plt.title("Logistic map bifurcation diagram")
plt.xlabel("r")
plt.ylabel("x*")
plt.show()
```
